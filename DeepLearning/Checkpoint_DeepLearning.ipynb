# 1. Checkpoint
Training of Deep learning can take hours and suddenly stop and lose a lot of work. One way it to use checkpoint to 
obtain the history of training. Even to obtain the best model. Application checkpoints is a fault tolerance technique for a long run process. Keras provides ModelCheckpoint callback class todo the job.

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
import numpy as np
seed =7
np.random.seed(seed)
```

```python
dataset = np.loadtxt("/home/tri/Downloads/MLdatasets/pima-indians-diabetes.data",delimiter=",")
X = dataset[:,0:8]
y= dataset[:,8]
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(12, input_dim=8,kernel_initializer='uniform',activation="relu"))
    model.add(Dense(8, kernel_initializer='uniform',activation='relu'))
    model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
# checkpoint
model = create_model()
filepath="weight-improvement-{epoch:02d}-{val_acc:.2f}.hdf5"
checkpoint= ModelCheckpoint(filepath, monitor='val_acc',save_best_only=True,mode='max',verbose=1)
callbacks_list =[checkpoint]
# fit model
#model.fit(X,y,validation_split=0.33, epochs=150, batch_size=10, callbacks=callbacks_list,verbose=0)
model.fit(X,y,validation_split=0.33,batch_size=10,epochs=100,callbacks=callbacks_list)
```

## Usage: using checkpoint best neural network model only

```python
from keras.callbacks import ModelCheckpoint
```

```python
# make check point
filepath ="weights.best.hdf5"
checkpoint= ModelCheckpoint(filepath, monitor='val_acc',verbose=1, save_best_only=True,mode='max')
callbacks_list =[checkpoint]
model.fit(X,y,validation_split=0.33,batch_size=10,epochs=100,callbacks=callbacks_list,verbose=0)
```

## Loading a saved neural model

```python
# load weights
model.load_weights("weights.best.hdf5")
```

```python
# compile model before predicting
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
# Load data
dataset = np.loadtxt("/home/tri/Downloads/MLdatasets/pima-indians-diabetes.data",delimiter=",")
X = dataset[:,0:8]
y= dataset[:,8]
```

```python
# estimating accuracy
scores =model.evaluate(X,y,verbose=0)
print("%s: %.2f%%" %(model.metrics_names[1],scores[1]*100))
```
